# Reading and Loading Data

```python
import numpy as np
```

```python
line_num = 0

X = []

for line in open("AirPassengers.csv"):
    if line_num > 0:
        row = line.split(',')[1:]
        sample = map(float, row)
        print(sample)
    
    line_num += 1
    X.append(sample)
```

We next can take this list of lists and convert it into a two-dimensional Numpy array:

```python
X = np.array(X)
X.shape
```

You can also import data using `pandas` by creating `dataframe` objects.
